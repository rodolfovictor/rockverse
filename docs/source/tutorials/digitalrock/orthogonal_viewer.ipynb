Exploring the orthogonal viewer
===============================

The Orthogonal Viewer is a convenient tool for making quick vilualizations of 3D voxel images.
In this tutorial, we'll explore how to use the ``Orthogonal Viewer`` class.

First let's open the Belgian Fieldstone sample we [imported before](./sample_data/import_belgian_fieldstone.ipynb):

```python
import rockverse as rv
belgian_fieldstone = rv.open(store='/path/to/imported/BelgianFieldstone/original')
belgian_fieldstone.info
```

```python
belgian_fieldstone.save('/path/to/imported/BelgianFieldstone/original2', chunks=(550,550,100), overwrite=True)
image2=rv.open('/path/to/imported/BelgianFieldstone/original2')
image2.info
```

```python
import matplotlib.pyplot as plt
plt.ion()

viewer = rv.digitalrock.OrthogonalViewer(image=image2)
```
